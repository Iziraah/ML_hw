```python
# Поиск выбросов с помощью метода квартилей
Q1 = np.percentile(X_train_numeric['Annual Income (k$)'], 25)
Q3 = np.percentile(X_train_numeric['Annual Income (k$)'], 75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers1 = [x for x in X_train_numeric['Annual Income (k$)'] if x < lower_bound or x > upper_bound]

Q1 = np.percentile(X_train_numeric['Age'], 25)
Q3 = np.percentile(X_train_numeric['Age'], 75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers2 = [x for x in X_train_numeric['Age'] if x < lower_bound or x > upper_bound]
```

### Задание 6
Нормализуйте численные признаки. Аргументируйте выбор стратегии нормализации для каждого признака.

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler
```

```python
# Создание экземпляров нормализаторов
min_max_scaler = MinMaxScaler()
standard_scaler = StandardScaler()
```

```python
# Нормализация признаков
X_train_numeric['Age'] = min_max_scaler.fit_transform(X_train[['Age']])
X_train_numeric['Annual Income (k$)'] = standard_scaler.fit_transform(X_train_numeric[['Annual Income (k$)']])
```

- Стратегия: Min-Max Scaling
Обоснование: Возраст измеряется на линейной шкале и не содержит выбросов, поэтому важно сохранить относительные различия между значениями. Применение Min-Max Scaling масштабирует значения в диапазон [0, 1], при этом относительные расстояния между возрастами остаются неизменными.

- Стратегия: Z-Score Standardization
Обоснование: Годовой доход имеет линейную шкалу, но может включать выбросы (например, покупатели с очень высоким доходом). Z-Score Standardization нормализует данные, центрируя их вокруг среднего и масштабируя относительно стандартного отклонения, что снижает влияние выбросов и способствует получению нормального распределения.

В данном примере использованы разные методы нормализации для каждого числового признака в зависимости от их характеристик. Такой дифференцированный подход позволяет повысить эффективность моделей машинного обучения и улучшить точность результатов.
